```python
from pathlib import Path
import mne_bids
import mne

bids_root = '/Volumes/T5_EVO/REPLAYSEQ/ICM/Data_ICM/BIDS'
bp = mne_bids.BIDSPath(
    root=bids_root,
    subject='01',
    task='reproduction',
    run='01',
    suffix='meg',
    extension='.fif',
    datatype='meg',
)
raw = mne_bids.read_raw_bids(bp)
if 'eeg' in raw:
    raw.set_montage('easycap-M1')
    raw.get_montage().plot(
        sphere='eeglab'  # new in MNE 1.1
    )
else:
    print("No EEG channels found in the data.")
```
